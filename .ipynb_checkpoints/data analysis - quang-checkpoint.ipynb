```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
dataframes = []

for i in range(1,6):
    df = pd.read_csv(f'data/channels_day_{i}.csv')
    dataframes.append(df)
    
channels_df = pd.concat(dataframes, ignore_index=True)
channels_df = channels_df.drop_duplicates(subset='Channel ID', keep='last')
channels_df
```

```python
videos_df = pd.read_csv(f'data/dataNew.csv')
videos_df = videos_df.drop('Unnamed: 0', axis=1)
videos_df.head()
videos_df.shape
```

```python
#Lấy videos xuất hiện hơn 15 lần
value_counts = videos_df['Video ID'].value_counts()
videos_df = videos_df[videos_df['Video ID'].isin(value_counts[value_counts >= 15].index)]

videos_df.shape
```

```python
unique_values = videos_df['Channel ID'].unique()

# lọc để dataframe channel và dataframe videos có chung các channel ID
channels_df = channels_df[channels_df['Channel ID'].isin(unique_values)]
print(channels_df.shape)
videos_df = videos_df[videos_df['Channel ID'].isin(unique_values)]
print(videos_df.shape)
```

```python
merged_df = pd.merge(videos_df, channels_df, on='Channel ID', how='inner')
merged_df
```

```python
print(merged_df.groupby('Video ID').size())

is_large_group = any(merged_df.groupby('Video ID').size() > 33)
if is_large_group:
    print("There is at least one group with a size greater than 33.")
else:
    print("No group has a size greater than 33.")
```

```python
print(videos_df.groupby('Channel ID').size())
```

```python
# chia channels thanh 3 phan khuc tuong ung voi q1, q2, q3 trong tu phan vi
q1_subscriber = np.percentile(channels_df['Num Subscriber'], 25, interpolation = 'linear')
q2_subscriber = np.percentile(channels_df['Num Subscriber'], 50, interpolation = 'linear')
q3_subscriber = np.percentile(channels_df['Num Subscriber'], 75, interpolation = 'linear')

# nhung channels co luot subscriber thap
low_subscriber_channels = channels_df[channels_df['Num Subscriber'] <= q1_subscriber]
low_subscriber_channels = low_subscriber_channels['Channel ID'].tolist()

# nhung channels co luot subscriber trung binh
medium_subscriber_channels = channels_df[(channels_df['Num Subscriber'] > q1_subscriber) & (channels_df['Num Subscriber'] < q3_subscriber)]
medium_subscriber_channels = medium_subscriber_channels['Channel ID'].tolist()

# nhung channels co luot subsriber cao
high_subscriber_channels = channels_df[channels_df['Num Subscriber'] >= q3_subscriber]
high_subscriber_channels = high_subscriber_channels['Channel ID'].tolist()

print(len(low_subscriber_channels))
print(len(medium_subscriber_channels))
print(len(high_subscriber_channels))
```

```python
def get_segment_high_channels(videos_df, high_subcriber_channels):
    segment_videos_high_channel = {}
    # get all videos in high subscriber channels
    videos_high_channel = []
    view_high_channel = []
    for channel_id in high_subcriber_channels:
        corresponding_video_df = videos_df[videos_df['Channel ID'] == channel_id]
        videos = corresponding_video_df['Video ID'].tolist()
        videos_high_channel = videos_high_channel + videos

        view = corresponding_video_df['View'].tolist()
        view_high_channel = view_high_channel + view

    q1_view = np.percentile(view_high_channel, 25)
    q3_view = np.percentile(view_high_channel, 75)

    high_view_videos_all_channels = []
    medium_view_videos_all_channels = []
    low_view_videos_all_channels = []
    for channel_id in high_subcriber_channels:
        corresponding_video_df = videos_df[videos_df['Channel ID'] == channel_id]
        # get high view videos
        high_view_videos = corresponding_video_df[corresponding_video_df['View'] >= q3_view]
        high_view_videos = high_view_videos['Video ID'].tolist()
        high_view_videos_all_channels = high_view_videos_all_channels + high_view_videos

        # get medium view videos
        medium_view_videos = corresponding_video_df[(corresponding_video_df['View'] > q1_view) & (corresponding_video_df['View'] < q3_view)]
        medium_view_videos = medium_view_videos['Video ID'].tolist()
        medium_view_videos_all_channels = medium_view_videos_all_channels + medium_view_videos
        
        # get low view videos
        low_view_videos = corresponding_video_df[corresponding_video_df['View'] <= q1_view]
        low_view_videos = low_view_videos['Video ID'].tolist()
        low_view_videos_all_channels = low_view_videos_all_channels + low_view_videos

    segment_videos_high_channel['Top 1'] = high_view_videos_all_channels
    segment_videos_high_channel['Top 2'] = medium_view_videos_all_channels
    segment_videos_high_channel['Top 3'] = low_view_videos_all_channels
    return segment_videos_high_channel
```

```python
def get_segment_medium_channels(videos_df, medium_subcriber_channels):
    segment_videos_medium_channel = {}
    # get all videos in high subscriber channels
    videos_medium_channel = []
    view_medium_channel = []
    for channel_id in medium_subcriber_channels:
        corresponding_video_df = videos_df[videos_df['Channel ID'] == channel_id]
        videos = corresponding_video_df['Video ID'].tolist()
        videos_medium_channel = videos_medium_channel + videos

        view = corresponding_video_df['View'].tolist()
        view_medium_channel = view_medium_channel + view

    q1_view = np.percentile(view_medium_channel, 25)
    q3_view = np.percentile(view_medium_channel, 75)

    high_view_videos_all_channels = []
    medium_view_videos_all_channels = []
    low_view_videos_all_channels = []
    for channel_id in medium_subcriber_channels:
        corresponding_video_df = videos_df[videos_df['Channel ID'] == channel_id]
        # get high view videos
        high_view_videos = corresponding_video_df[corresponding_video_df['View'] >= q3_view]
        high_view_videos = high_view_videos['Video ID'].tolist()
        high_view_videos_all_channels = high_view_videos_all_channels + high_view_videos

        # get medium view videos
        medium_view_videos = corresponding_video_df[(corresponding_video_df['View'] > q1_view) & (corresponding_video_df['View'] < q3_view)]
        medium_view_videos = medium_view_videos['Video ID'].tolist()
        medium_view_videos_all_channels = medium_view_videos_all_channels + medium_view_videos
        
        # get low view videos
        low_view_videos = corresponding_video_df[corresponding_video_df['View'] <= q1_view]
        low_view_videos = low_view_videos['Video ID'].tolist()
        low_view_videos_all_channels = low_view_videos_all_channels + low_view_videos

    segment_videos_medium_channel['Top 1'] = high_view_videos_all_channels
    segment_videos_medium_channel['Top 2'] = medium_view_videos_all_channels
    segment_videos_medium_channel['Top 3'] = low_view_videos_all_channels
    return segment_videos_medium_channel
```

```python
def get_segment_low_channels(videos_df, low_subcriber_channels):
    segment_videos_low_channel = {}
    # get all videos in high subscriber channels
    videos_low_channel = []
    view_low_channel = []
    for channel_id in low_subcriber_channels:
        corresponding_video_df = videos_df[videos_df['Channel ID'] == channel_id]
        videos = corresponding_video_df['Video ID'].tolist()
        videos_low_channel = videos_low_channel + videos

        view = corresponding_video_df['View'].tolist()
        view_low_channel = view_low_channel + view

    q1_view = np.percentile(view_low_channel, 25)
    q3_view = np.percentile(view_low_channel, 75)

    high_view_videos_all_channels = []
    medium_view_videos_all_channels = []
    low_view_videos_all_channels = []
    for channel_id in low_subcriber_channels:
        corresponding_video_df = videos_df[videos_df['Channel ID'] == channel_id]
        # get high view videos
        high_view_videos = corresponding_video_df[corresponding_video_df['View'] >= q3_view]
        high_view_videos = high_view_videos['Video ID'].tolist()
        high_view_videos_all_channels = high_view_videos_all_channels + high_view_videos

        # get medium view videos
        medium_view_videos = corresponding_video_df[(corresponding_video_df['View'] > q1_view) & (corresponding_video_df['View'] < q3_view)]
        medium_view_videos = medium_view_videos['Video ID'].tolist()
        medium_view_videos_all_channels = medium_view_videos_all_channels + medium_view_videos
        
        # get low view videos
        low_view_videos = corresponding_video_df[corresponding_video_df['View'] <= q1_view]
        low_view_videos = low_view_videos['Video ID'].tolist()
        low_view_videos_all_channels = low_view_videos_all_channels + low_view_videos

    segment_videos_low_channel['Top 1'] = high_view_videos_all_channels
    segment_videos_low_channel['Top 2'] = medium_view_videos_all_channels
    segment_videos_low_channel['Top 3'] = low_view_videos_all_channels
    return segment_videos_low_channel
```

```python
# doi voi moi phan khuc channel tiep tuc chia phan khuc video tuong ung
segment_videos_high_channel = get_segment_high_channels(videos_df, high_subcriber_channels)
segment_videos_medium_channel = get_segment_low_channels(videos_df, medium_subcriber_channels)
segment_videos_low_channel = get_segment_low_channels(videos_df, low_subscriber_channels)

print(segment_videos_high_channel)
print(segment_videos_medium_channel)
print(segment_videos_low_channel)
```

## Analyze channel segment based on 'Num Subscribers' attribute

```python
#thêm cột phân khúc vào dataframe channels
channels_df['segment'] = channels_df['Channel ID'].apply(lambda x: 'low' if x in low_subscriber_channels 
                                                       else ('medium' if x in medium_subscriber_channels 
                                                       else 'high'))
std_subs_segments = channels_df.groupby('segment')['Num Subscriber'].std().reset_index(name='std_value')
mean_subs_segments = channels_df.groupby('segment')['Num Subscriber'].mean().reset_index(name='mean_value')
median_subs_segments = channels_df.groupby('segment')['Num Subscriber'].median().reset_index(name='median_value')

print(mean_subs_segments)
print(median_subs_segments)
print(std_subs_segments)
```

```python
# Set the figure size
plt.figure(figsize=(11, 4.5))

# bar chart for mean
bars = plt.barh(mean_subs_segments['segment'], mean_subs_segments['mean_value'])
plt.xlabel('Segment')
plt.ylabel('Mean of Subscribers')
plt.title('Mean number of subscribers per segment')
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, format(width, '.2f'), ha='left', va='center')
plt.show()

# bar chart for median
plt.figure(figsize=(11, 4.5))
bars_median = plt.barh(median_subs_segments['segment'], median_subs_segments['median_value'])
plt.xlabel('Segment')
plt.ylabel('Median of Subscribers')
plt.title('Median number of subscribers per segment')
for bar in bars_median:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, format(width, '.2f'), ha='left', va='center')
plt.show()

# bar chart for standard deviation
plt.figure(figsize=(11, 4.5))
bars_std = plt.barh(std_subs_segments['segment'], std_subs_segments['std_value'])
plt.xlabel('Segment')
plt.ylabel('Standard Deviation of Subscribers')
plt.title('Standard deviation of subscribers per segment')
for bar in bars_std:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, format(width, '.2f'), ha='left', va='center')
plt.show()
```

#### Comments on 3 segments based on the number of subsrcibers:
- Low segment:
  - This segment only has 13239.57 subscribers, which is almost 10 times smaller than medium segment and 300 times smaller than the high segment. The low segment also has low standard deviation, which suggest relatively consistent subscriber counts across channels in this segment
- medium segment:
  - This segment has the average number of subscribers approximately 119,000 and a moderate standard deviation about 75000.
- high segment:
  - This segment has an outstanding mean number of subsrcibers, about 3 million, but it also has great standard deviation with almost 7 million, indicating significant variability in the number of subscribers across channels in this segment.

## Analyze videos segment based on View attribute

```python
# Tạo cột mới với giá trị ban đầu là None
videos_df['vid_segment'] = None
videos_df['channel_segment'] = None
# Duyệt qua các tuple
for i, (key, value) in enumerate(segment_videos_high_channel.items()):
    # Ánh xạ và gán giá trị từ tuple1 vào cột mới dựa trên ID
    videos_df.loc[videos_df['Video ID'].isin(value), 'vid_segment'] = f'top {i+1}'
    videos_df.loc[videos_df['Video ID'].isin(value), 'channel_segment'] = f'high'

for i, (key, value) in enumerate(segment_videos_medium_channel.items()):
    # Ánh xạ và gán giá trị từ tuple2 vào cột mới dựa trên ID
    videos_df.loc[videos_df['Video ID'].isin(value), 'vid_segment'] = f'top {i+1}'
    videos_df.loc[videos_df['Video ID'].isin(value), 'channel_segment'] = f'medium'

for i, (key, value) in enumerate(segment_videos_low_channel.items()):
    # Ánh xạ và gán giá trị từ tuple3 vào cột mới dựa trên ID
    videos_df.loc[videos_df['Video ID'].isin(value), 'vid_segment'] = f'top {i+1}'
    videos_df.loc[videos_df['Video ID'].isin(value), 'channel_segment'] = f'low'
```

```python
mean_view_segments = videos_df.groupby(['channel_segment', 'vid_segment'])['View'].mean().reset_index(name='mean_value')
median_view_segments = videos_df.groupby(['channel_segment', 'vid_segment'])['View'].median().reset_index(name='median_value')
std_view_segments = videos_df.groupby(['channel_segment', 'vid_segment'])['View'].std().reset_index(name='std_value')

print(mean_view_segments)
print(median_view_segments)
print(std_view_segments)
```

### Analyze Videos from Low channels segment

```python
std_view_segments_low = std_view_segments[std_view_segments['channel_segment']=='low']
mean_view_segments_low = mean_view_segments[mean_view_segments['channel_segment']=='low']
median_view_segments_low = median_view_segments[median_view_segments['channel_segment']=='low']

# Create the figure and subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
# Plotting the subplots
axes[0].bar(mean_view_segments_low['vid_segment'], mean_view_segments_low['mean_value'])
axes[0].set_title("Mean - Low channel segment")
axes[0].set_xlabel('Video segment')
axes[0].set_ylabel('Mean Views')

axes[1].bar(std_view_segments_low['vid_segment'], std_view_segments_low['std_value'])
axes[1].set_title("Standard deviation - Low channel segment")
axes[1].set_xlabel('Video segment')
axes[1].set_ylabel("Standard deviation of Views")

axes[2].bar(median_view_segments_low['vid_segment'], median_view_segments_low['median_value'])
axes[2].set_title("Median - Low channel segment")
axes[2].set_xlabel('Video segment')
axes[2].set_ylabel('Median Views')

# Adjust spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```
